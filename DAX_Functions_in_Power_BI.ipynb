**DAX Functions in Power BI**

**Q1.Create a calculated column TotalPrice using Quantity and Unit Price.**

**ANS:**

* Open Power BI Desktop.
* Get Data -> Csv -> Load

* Go to Data view.

* Click New Column.

* Enter the DAX formula:

* Total Sales = amazon_store_sales_2022[Quantity]* amazon_store_sales_2022[UnitPrice]

* Click Commit.

* Verify that a new column appears with calculated values.
